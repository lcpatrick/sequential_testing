```python
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
sigma = [[1, -0.7], [-0.7, 1]]
mean = np.array([1, 2])
N = 1000
sims = 10000
Z = np.zeros((sims, 2))
for i in range(sims):
    control = np.random.multivariate_normal(mean, sigma, N)
    treatment = np.random.multivariate_normal(mean, sigma, N)
    z_0 = (control[:, 0].mean() - treatment[:, 0].mean())/np.sqrt(2 / sims)
    z_1 = (control[:, 1].mean() - treatment[:, 1].mean())/np.sqrt(2 / sims)
    Z[i, 0] = z_0
    Z[i, 1] = z_1
np.corrcoef(Z[:, 0], Z[:, 1])
```

```python
sns.distplot(Z[:, 0])
```

```python
sns.distplot(Z[:, 1])
```
